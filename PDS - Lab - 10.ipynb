# Generate scatter plot grouped by smoker field and 
# age on x axis and charges on y axis
df = pd.read_csv('insurance.csv')

```python
df=pd.read_csv('insurance.csv')
group=df.groupby(['smoker'])
for k, group in group:
    plt.scatter(group['age'], group['charges'], label='Smoke='+k)
plt.legend()
plt.show()
```
