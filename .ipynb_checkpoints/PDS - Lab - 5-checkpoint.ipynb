# 1)	Scrape the data of Faculty and news data from darshan university website.



```python
import requests
import bs4
data = requests.get("https://www.darshan.ac.in/engineering/computer/faculty")
soap = bs4.BeautifulSoup(data.text,'lxml')
allfaculty = soap.select("body > main > div:nth-child(5) > div > div > div > div > div")

for f in allfaculty:
    name = f.select("h2")[0].text.strip()
    post = f.select("span")[0].text.strip()
    education = f.select("span")[1].text.strip()
    experience = f.select("span")[3].text.strip()
    working_since = f.select("span")[5].text.strip()
    
    print("Name : ",name)
    print("Post : ",post)
    print("Eduction : ",education)
    print("Experience : ",experience)
    print("Working since : ",working_since)
    print("\n")
```

# 2) Scrape the data of student placement from darshan university website.

```python
import requests
import bs4
data = requests.get("https://www.darshan.ac.in/placement/list/engineering/computer/2023")
soap = bs4.BeautifulSoup(data.text,'lxml')
allstudent = soap.select("body > main > div:nth-child(5) > div > div > div.col-lg-8.col-xl-9.g-font-size-14.g-mb-30 > div.g-pt-20 > div.row.g-px-4.g-px-5--sm.g-px-5--md >div")

for s in allstudent:
    name = s.select("h2")[0].text.strip()
    enrollmentNo = s.select("p")[0].text.strip()
    placedCity = s.select("p")[4].text.strip()
    print(name,"--", enrollmentNo, "--", placedCity)
```
